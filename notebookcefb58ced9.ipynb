```python
x_train = np.array(x_train)
y_train = np.array(y_train)
```

```python
# Normalize the image data to [0, 1]
x_train = x_train / 255.0
```

```python
import numpy as np

# Convert y_train to one-hot encoded matrix
num_classes = 5
y_train_encoded = np.zeros((len(y_train), num_classes))
y_train_encoded[np.arange(len(y_train)), y_train] = 1

# Print the shape of y_train_encoded
print("y_train_encoded shape:", y_train_encoded.shape)
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x_train, y_train_encoded, test_size=0.15, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=60,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='constant'
)


def augment_dataset(dataset):
    augmented_data = []
    for img in dataset:
        img = np.expand_dims(img, axis=0)  
        for _ in range(2):  
            seed = np.random.randint(0, 1000)  
            for batch in datagen.flow(img, batch_size=1, seed=seed):
                augmented_data.append(np.squeeze(batch))  
                break  
    return np.array(augmented_data)
x_train_augmented = augment_dataset(X_train)
y_train_augmented = np.repeat(y_train,2, axis=0)  
```

```python
print(x_train_augmented.shape)
print(y_train_augmented.shape)

print(X_test.shape)
print(y_test.shape)
```
